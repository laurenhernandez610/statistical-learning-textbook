```python
"""
Q1. If we list all the natural numbers below 10 that are multiples of 3 or 5,
we get 3, 5, 6 and 9. The sum of these multiples is 23.
Find the sum of all the multiples of 3 or 5 below 1000.
"""
def function(x1 = 3, x2 = 5, number = 1000):
    """
    This function will return all multiples of x1 and x2 below a given number
    inputs(x1, x2, number): int
    output: sum of all multiples of x1 and x2
    """
    final_sum = 0
    for i in range(number):
        if i % x1 == 0 or i % x2 == 0:
            final_sum += i
    return final_sum
function()
```
